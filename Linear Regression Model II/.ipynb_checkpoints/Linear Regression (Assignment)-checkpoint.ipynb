# Using the formular for Linear Regression

```python
Transpose = X.T
Transpose
```

```python
matrix_X = np.dot(Transpose, X)
matrix_X
```

```python
matrix_inverse = np.linalg.inv(matrix_X)
matrix_inverse
```

```python
matrix_3 = np.dot(matrix_inverse, Transpose)
matrix_3
```

```python
matrix_ans_1D = np.dot(matrix_3, Y_1D)
matrix_ans_1D
```

```python
matrix_ans = np.dot(matrix_3, Y)
matrix_ans
```

# Substituting Answers in Formular

```python
data
```

```python
# Where Y(lINEAR) = W1X1 + W2X2 + W3X3 + W4X4 + B
W1 = 0.67116016
W2 = 0.00552638
W3 =-0.38099038
W4 = 0.94470692
B = -1.89357902
```

```python
Y_0 = (W1*4 + W2*85 + W3*6 + W4*70 + B)
Y_0
```

```python
Y_1 = (W1*6 + W2*90 + W3*7 + W4*80 + B)
Y_1
```

```python
Y_LINEAR = (W1*data.iloc[:,0] + W2*data.iloc[:,1] + W3*data.iloc[:,2]+ W4*data.iloc[:,3] + B )
Y_LINEAR
```

```python
Predicted_score = Y_LINEAR 
```

```python
data_final = data.copy()
```

```python
data_final['Predicted_score'] = Predicted_score
data_final
```

```python
Error = data_final['Final_Score'] - data_final['Predicted_score']
```

```python
rms = np.sqrt(np.mean(np.square(Error)))
print("The calculated Root Mean Square is:", rms)
```

```python
data_final['Predicted_score(rms)'] = data_final['Predicted_score'] - rms
```

```python
data_final
```

```python
data_final.astype(int)
```

```python
# From this analysis, the RMS is not needed as the predicted score is very close in comparison, to the final score, so we can drop the adjusted score.
data_f = data_final.drop(columns = ['Predicted_score(rms)'])
data_f
```

```python
data_f.to_csv('ML_score.csv', index = False)
```
