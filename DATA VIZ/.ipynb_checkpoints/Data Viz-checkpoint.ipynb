```python
# Pie chart cannot work for large datasets
```

```python
months = ["January", "February", "March", "April", "June", "July"]
returns = [40, 30, 60, 70, 80, 40]
```

```python
dataset = pd.DataFrame({'Month':months, 'Returns': returns})
```

```python
dataset.reset_index(drop=True)
```

```python
dataset = dataset.reset_index(drop=True)
```

```python
new_index = pd.RangeIndex(start=1, stop=len(dataset) + 1, step=1)
```

```python
# Assign the new index to the DataFrame
dataset.index = new_index
dataset
```

```python
# Creating a pie chart
```

```python
plt.pie(returns, labels = months, autopct = '%1.1f%%', startangle = 90)
plt.title("months vs returns")
plt.axis("equal")
plt.show()
```

```python
#Creating a bar chart
```

```python
#Crete the bar chart
plt.figure(figsize = (6,3)) #set the figure size
plt.bar(months, returns, color = 'red')

#Add title and labels
plt.title("Simple bar chart")
plt.xlabel("months")
plt.ylabel("returns")

#Show the plots
plt.show()
```

```python
# Creating Scattered Plots
```

```python
plt.figure(figsize = (5,3))
plt.scatter(months, returns)
plt.xlabel("months")
plt.ylabel("returns")
plt.title("My Scatter Plot")
plt.show()
```

```python
plt.figure(figsize = (5,3))
plt.plot(months, returns)
plt.xlabel("months")
plt.ylabel("returns")
plt.title("My Scatter Plot")
plt.show()
```

```python
plt.figure(figsize = (6,3))
plt.hist(returns, bins = 5, edgecolor='black', alpha=0.7, color='lightcoral')
plt.xlabel("months")
plt.ylabel("returns")
plt.title("My histogram Plot")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Okay, if you specifically want to plot a histogram using your provided returns data, 
# you must understand that a histogram will show the distribution of the numerical returns values, 
# not a return value per month. The x-axis will represent ranges of return values, 
# and the y-axis will represent how many months fall into each of those ranges. 
# The months data will not be directly used as the bins or the primary x-axis labels for a histogram.
```

```python
# Why a bar chart is correct here:
# Categorical X-axis: months are distinct categories.
# Numerical Y-axis: returns are the values associated with each category.
# A bar chart directly shows the height (return) for each category (month).

# Why plt.hist() is not suitable here:
# plt.hist() is for showing the distribution of a single continuous numerical variable. 
# It bins the numerical data and shows how many data points fall into each bin.
# You cannot directly input categories like months into plt.hist() as the data to be binned.
```
