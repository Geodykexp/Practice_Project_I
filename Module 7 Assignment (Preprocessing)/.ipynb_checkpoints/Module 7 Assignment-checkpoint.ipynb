# Label Encoding

```python
import pandas as pd
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
car_data = pd.read_csv("car_price.csv")
```

```python
car_data.head()
```

```python
encoder = LabelEncoder()
```

```python
car_data.dropna()
```

```python
car_data["transmission"]
```

```python
car_data["transmission"] = encoder.fit_transform(car_data["transmission"])
```

```python
car_data["transmission"]
```

```python
car_data.head()
```

```python
car_data["fuel_type"]
```

```python
car_data["fuel_type"] = encoder.fit_transform(car_data["fuel_type"])
```

```python
car_data["fuel_type"]
```

```python
car_data.head()
```

```python
car_data.info()
```

```python
car_data.dropna(inplace = True)
```

```python
car_data.isnull().sum()
```

```python
car_data["engine"] = car_data["engine"].str.replace("cc", "")
```

```python
car_data["kms_driven"] = car_data["kms_driven"].str.replace(",", "")
```

```python
def Grading(x):
    if x > 200:
        print("Big")
    elif x > 100 and x <= 200:
        print("Average")
    else:
        print("Small")
```

```python
print(Grading(300))
```
