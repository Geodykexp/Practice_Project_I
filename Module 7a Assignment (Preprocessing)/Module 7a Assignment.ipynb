```python
def square_feet_converter(value):
    try:
        value = value.strip().split()
        if value[1] == "sqft":
            value[0] = float(value[0])
            converted_value = value[0] * 1
        if value[1] == "sqm":
            value[0] = float(value[0])
            converted_value = value[0] * 10.7639
        if value [1] == "sqyrd":
            value[0] = float(value[0])
            converted_value = value[0] * 9
        if value [1] == "rood":
            value[0] = float(value[0])
            converted_value = value[0] * 10890
        return converted_value
    except:
        return value
            
```

```python
dataset["square_feet"] = dataset["square_feet"].apply(square_feet_converter)
```

```python
def price_converter(value):
    try:
        value = value.strip().split()
        if value[1] == "Lac":
            value[0] = float(value[0])
            converted_value = value[0] * 100000
        if value[1] == "Cr":
            value[0] = float(value[0])
            converted_value = value[0] * 10000000
        return converted_value
    except:
        return value        
    
```

```python
dataset["price"] = dataset["price"].apply(price_converter)
```
